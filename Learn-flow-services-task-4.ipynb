## Step 3: Exploratory Data Analysis (EDA)

```python
# Basic statistics of numerical columns
print(data.describe())

# Visualize the distribution of age
data['age'].hist(bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Import necessary libraries for visualization
import seaborn as sns

# Visualize the distribution of 'friend_count'
plt.figure(figsize=(10, 6))
sns.histplot(data['friend_count'], bins=30, kde=True)
plt.title('Distribution of Friend Count')
plt.xlabel('Friend Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot to identify outliers in 'friend_count'
plt.figure(figsize=(8, 5))
sns.boxplot(x=data['friend_count'])
plt.title('Boxplot of Friend Count')
plt.xlabel('Friend Count')
plt.show()
```

```python
# Scatter plot to explore the relationship between 'friend_count' and 'likes_received'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='friend_count', y='likes_received', data=data, alpha=0.5)
plt.title('Scatter Plot: Friend Count vs Likes Received')
plt.xlabel('Friend Count')
plt.ylabel('Likes Received')
plt.show()
```

```python
# Bar plot for gender distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', data=data, palette='viridis')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart for gender distribution
gender_counts = data['gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.show()
```

## Step 4: Identify Influential Users

```python
# Import necessary libraries
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph from the 'friend_count' column
G = nx.from_pandas_edgelist(data, 'userid', 'friend_count', create_using=nx.DiGraph())

# Calculate degree centrality
degree_centrality = nx.degree_centrality(G)

# Identify the most influential users
most_influential_users = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:10]
print("Top 10 Influential Users:", most_influential_users)
```
