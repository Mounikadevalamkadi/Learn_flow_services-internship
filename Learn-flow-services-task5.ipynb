# Data Preprocessing

 Identify Missing Values

```python
df.isnull().sum()
```

```python
# Drop rows with missing values for specific columns
df_cleaned = df.dropna(subset=['TweetID', 'Weekday', 'Hour', 'Day', 'Lang'])
# Display the shape of the cleaned DataFrame
print("Shape after handling missing values:", df_cleaned.shape)
```

```python
# Define numeric_columns based on the numeric columns in your DataFrame
numeric_columns = ['Hour', 'Day', 'IsReshare', 'Reach', 'RetweetCount', 'Likes', 'Klout', 'Sentiment', 'LocationID']
# Convert numeric columns to appropriate data types using .loc
df_cleaned.loc[:, numeric_columns] = df_cleaned[numeric_columns].apply(pd.to_numeric, errors='coerce')
# Display the data types after conversion
print("\nData Types after conversion:")
print(df_cleaned.dtypes)
# Check for missing values again
missing_values_cleaned = df_cleaned.isnull().sum()
print("\nMissing Values after cleaning:")
print(missing_values_cleaned)
```

```python
# Impute missing values for numeric columns using .loc
df_cleaned.loc[:, numeric_columns] = df_cleaned.loc[:, numeric_columns].apply(lambda x: x.fillna(x.mean()))
# Display the DataFrame after imputation
df_cleaned.head()
```

# Step 7: Sentiment Analysis

```python
import matplotlib.pyplot as plt

# Count sentiments
sentiment_counts = df['Sentiment'].value_counts()

# Plot sentiment distribution with stacked bar chart
plt.figure(figsize=(8, 5))
sentiment_counts.plot(kind='bar', color=['red', 'green', 'blue'])
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Analysis')
plt.show()
print(sentiment_counts)
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import seaborn as sns
import matplotlib.pyplot as plt

# Download the VADER lexicon for sentiment analysis
nltk.download('vader_lexicon')

# Initialize the SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Assuming 'text' is the column containing the tweets
df_cleaned['SentimentScore'] = df_cleaned['text'].dropna().apply(lambda x: sia.polarity_scores(x)['compound'])

# Visualize sentiment distribution
plt.figure(figsize=(10, 6))

# Check if 'SentimentScore' is present in the DataFrame
if 'SentimentScore' in df_cleaned.columns:
    sns.histplot(df_cleaned['SentimentScore'], bins=30, kde=True)
    plt.title('Distribution of Sentiment Scores in Tweets')
    plt.xlabel('Sentiment Score')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("Error: 'SentimentScore' column not found in the DataFrame.")
print(df_cleaned.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'Sentiment' is a categorical column in your DataFrame
# Replace 'Sentiment' with the actual column name representing categories
sns.pairplot(df_cleaned.sample(n=500), hue='Sentiment', diag_kind='kde', markers='.', palette='viridis')
plt.suptitle('Pairplot for Numeric Columns with Categorical Hue (Subsample)', y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation calculation
numeric_columns = df_cleaned.select_dtypes(include=['number']).columns
numeric_data = df_cleaned[numeric_columns]

# Compute the correlation matrix for numeric columns
correlation_matrix = numeric_data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Create a heatmap with the correlation values
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)

plt.title('Correlation Heatmap for Numeric Columns')
plt.show()
```

```python
# Scatter plots for numeric columns vs. engagement metrics
engagement_columns = ['Reach', 'RetweetCount', 'Likes']

plt.figure(figsize=(16, 12))

for i, engagement_metric in enumerate(engagement_columns):
    plt.subplot(3, 1, i + 1)
    
    # Select only numeric columns for the scatter plot
    numeric_columns_subset = df_cleaned.select_dtypes(include=['number'])
    
    # Scatter plot for each numeric column against the engagement metric
    for column in numeric_columns_subset.columns:
        sns.scatterplot(x=column, y=engagement_metric, data=df_cleaned)
    
    plt.title(f'Scatter Plots: {engagement_metric} vs. Numeric Columns')
    plt.xlabel('Numeric Columns')
    plt.ylabel(engagement_metric)

plt.tight_layout()
plt.show()
```
